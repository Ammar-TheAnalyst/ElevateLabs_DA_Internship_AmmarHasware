```python
df['Age'].fillna(df['Age'].median(), inplace=True)
df
```

# The dataset was containing 177 missing values in age column, so it was imputed using the median.

```python
df['Embarked'].fillna(df['Embarked'].mode()[0], inplace=True)
df.isnull().sum()
```

# Also it was having just 2 missing values in embarked column, so it was imputed using the mode

```python
df['cabin_new'] = df['Cabin'].notnull().astype(int)
df.drop('Cabin', axis=1, inplace=True)
df.isnull().sum()
```

# And at last, it was having 687 missing values in cabin column and its too much, so instead of imputing unreliable data, a new binary variable cabin_new was created and the original column was dropped.

```python
df['Survived'] = df['Survived'].apply(lambda x: 
                                      'Yes' if x == 1 
                                      else 
                                      'No')
```

# Converted 0/1 values of survived column into yes/no, so it would be understanable

```python
df.value_counts()
```

```python
df['Survived'].value_counts()
df['Sex'].value_counts()
df['Embarked'].value_counts()
```

```python
plt.figure(figsize=(6,5))
ax = sns.countplot(x='Survived', data=df, hue='Sex', palette={'male':'steelblue', 'female':'lightpink'})
for container in ax.containers:
    ax.bar_label(container)
plt.title("Count of Survived Passengers by Gender")
plt.show()
```

# From the countplot, we can observe that a higher number of females survived compared to males, while the majority of males did not survive.

```python
counts = df['Survived'].value_counts()
plt.pie(counts, labels=counts.index, autopct='%1.1f%%', colors=['lightblue', 'orange'])
plt.title("Percentage of Survived Passengers")
plt.show()
```

# By observing the pie chart, we can conclude that 38.4% of Passengers were survived.

```python
plt.figure(figsize=(10, 6))
sns.histplot(x = "Age", data = df, color="blue")
plt.title("Age Distribution of Passengers boarded Titanic")
plt.show()
```

# Looking at this histplot, we can see that the majority of passengers who boarded the Titanic were around 30 years old.

```python
plt.figure(figsize=(8,6))
sns.boxplot(x = "Pclass", y = "Fare", data = df)
plt.xlabel("Passenger Class")
plt.show()
```

# The boxplot shows that passengers in higher class (Passenger class 1) paid significantly higher fares than those in lower classes.

```python
sns.pairplot(df[["Age", "Fare", "SibSp", "Survived"]], hue = "Survived", palette = "bright")
plt.suptitle("Pairwise Relationships Age, Fare and Sibsp", y=1.02)
plt.show()
```

```python
# From the pairplot, we can observe the following:
# ​1) Fare is the Strongest Predictor: Passengers who paid high fares had a significantly higher survival rate (orange dots dominate the high-fare range).
# ​2) Age Range: There is very young children (Age < 5) showing a better survival chance than other age groups.
# 3) ​Family Size: Having one Sibling/Spouse correlates with a good survival rate. However, large family groups had a poor chance of surviving.
# 4) ​High Fare/Low Family: The most expensive tickets were usually bought by single travelers or pairs (SibSp 0 or 1).
```
